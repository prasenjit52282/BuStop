```python
import os
import glob
```

# Unnecessry Files

```python
acc_files=glob.glob('./Trails/*/*/ACC.csv')
gps_files=glob.glob('./Trails/*/*/GPS.csv')
sound_files=glob.glob('./Trails/*/*/Bus_SOUND_*.wav')
wifi_files=glob.glob('./Trails/*/*/Bus_WiFi_*.txt')
```

```python
def delete_file(file):
    try:
        os.remove(file)
        print('File Successfully deleted: '+file)
    except FileNotFoundError:
        print("File already deleted: "+file)
```

# Deleting them

```python
for f in acc_files:
    delete_file(f)

for f in gps_files:
    delete_file(f)
    
for f in sound_files:
    delete_file(f)

for f in wifi_files:
    delete_file(f)
```
